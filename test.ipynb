```python
from ultralytics import YOLO

# 모델 로드
model = YOLO("runs/detect/yolov8n_v8_50e31/weights/best.pt")

# 영상 예측 실행
results = model.predict(
    source="datasets/pothole_dataset_v8/train/images",  # 예측할 영상 경로
    imgsz=1280,
    show=True,  # 예측 결과를 화면에 표시
    hide_labels=True,  # 라벨 숨김
    save=True,  # 결과 저장
    project="runs/detect",  # 결과 저장 경로
    name="yolov8n_v8_50e_infer1280"  # 저장 폴더 이름
)

print("Inference complete. Results saved to:", results[0].save_dir)
```
